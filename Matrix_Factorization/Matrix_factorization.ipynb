Matrix Factorization with alternating least square optimization algorith.

```python
import os 
import pickle 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.utils import shuffle 
from datetime import datetime 
from sortedcontainers import SortedList 

# load data 
os.chdir(r'/home/qshan/Desktop/Repos/Big_Data_Projects/Recommender_System/Matrix_Factorization')
with open('../data/ml-20m/user_movie', 'rb') as f:
    user_movie = pickle.load(f)
 
with open('../data/ml-20m/movie_user', 'rb') as f:
    movie_user = pickle.load(f)
    
with open('../data/ml-20m/usermovie_rating', 'rb') as f:
    usermovie_rating = pickle.load(f)
    
with open('../data/ml-20m/usermovie_rating_test', 'rb') as f:
    usermovie_rating_test = pickle.load(f)
```

```python
#  number of users and movies 
N = np.max(list(user_movie.keys())) + 1 

M = max(np.max(list(movie_user.keys())), np.max([m_id for (_, m_id), _ in usermovie_rating_test.items()])) + 1

print('Max of users id, N =', N, '\n')
print('Max movies id M, =', M, '\n')
```

```python
# create a dictionary with userId as key, a tuple of all movies and corresponding ratings rated by userId as value 
print('Creating dictionaries...')
# dict for train set 
user_movierating = {}
for i, movies in user_movie.items():
    ratings = np.array([usermovie_rating[(i, j)] for j in movies])
    user_movierating[i] = (movies, ratings)

# dict with movieId as key, a tuple of uesrs and corresponding ratings who rated the movie as value 
movie_userrating = {}
for j, users in movie_user.items():
    ratings = np.array([usermovie_rating[(i, j)] for i in users])
    movie_userrating[j] = (users, ratings)

# dict for test set 
movie_userrating_test = {}
for (i, j), rating in usermovie_rating_test.items():
    # if movie j not exists in current dict 
    if j not in movie_userrating_test:
        movie_userrating_test[j] = [[i], [rating]]
    else:
        movie_userrating_test[j][0].append(i)
        movie_userrating_test[j][1].append(rating)

# convert the ratings list to np.array 
for j, (users, ratings) in movie_userrating_test.items():
    movie_userrating_test[j][1] = np.array(ratings)
print('Dictionaries created')
```

```python
# initialization
# column number of matrix W, U
K = 10 
W = np.random.randn(N, K)
# user bias 
b = np.zeros(N)
U = np.random.randn(M, K)
# item bias 
c = np.zeros(M)
# global average 
mu = np.mean(list(usermovie_rating.values()))

# loss function 
def compute_loss(m2u):
    # total number of predictions 
    N = 0 
    sse = 0 
    for m, (u, r) in m2u.items():
        prediction = W[u].dot(U[m]) + b[u] + c[m] + mu
        residual = prediction - r
        sse += residual.dot(residual)
        N += len(r)
    return sse / N
```

```python
# model training 
epochs = 30 
lambda_ = 20 # penalty 
train_loss = []
test_loss = []

# loop 
for epoch in range(epochs):
    print('epoch: ', epoch)
    # record time for reference 
    epoch_start = datetime.now()
    
    # update W and b (over each user)
    for i in range(N):
        m_ids, ratings = user_movierating[i]
        # np.eye : matrix of 1s 
        matrix = U[m_ids].T.dot(U[m_ids]) + np.eye(K) * lambda_
        vector = (ratings - b[i] - c[m_ids] - mu).dot(U[m_ids])
        bi_sum_term = (ratings - U[m_ids].dot(W[i]) - c[m_ids] - mu).sum()
        
        # update 
        W[i] = np.linalg.solve(matrix, vector)
        b[i] = bi_sum_term / (len(user_movie[i]) + lambda_)
        
    # update U and c over all movies 
    for j in range(M):
        try:
            u_ids, ratings = movie_userrating[j]
            matrix = W[u_ids].T.dot(W[u_ids]) + np.eye(K) * lambda_
            vector = (ratings - b[u_ids] - c[j] - mu).dot(W[u_ids])
            cj = (ratings - W[u_ids].dot(U[j]) - b[u_ids] - mu).sum()
                
            # set the updates 
            U[j] = np.linalg.solve(matrix, vector)
            c[j] = cj / (len(movie_user[j]) + lambda_)
        except KeyError:
            # if a movie has no ratings from any user at all 
            pass
    print('epoch duration: ', datetime.now() - epoch_start)
        
    # train loss for each epoch 
    train_loss.append(compute_loss(movie_userrating))
    # test loss 
    test_loss.append(compute_loss(movie_userrating_test))
    print('train loss at epoch ', epoch, ':', train_loss[-1])
    print('test loss at epoch ', epoch, ':', test_loss[-1])

        
        
```

```python
# plot the loss 
plt.plot(train_loss, label = 'train loss')
plt.plot(test_loss, label = 'test loss')
plt.title('Loss in MSE')
plt.legend()
plt.show()
```
