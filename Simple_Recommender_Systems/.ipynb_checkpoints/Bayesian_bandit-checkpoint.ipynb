## Bayesian Approach to Find Out the Best Rewarded Slot Machine 

* Model each slot machine as a separate beta distribution, which is going to model each slot machine's win rate.
* In each loop, draw a sample from each of beta distributions and sort them in order (ranking), and this is the order to rank items in a recommendation system.
* Pick the top item and play that machine only.


```python
from __future__ import print_function, division 

import matplotlib.pyplot as plt 
import numpy as np 
from scipy.stats import beta # beta distribution 
```

```python
# define constants: number of trials, true bandit probabilities  
NUM_TRIALS = 2000
BANDIT_PROBABILITIES = [0.2, 0.5, 0.75]
```

```python
class Bandit(object):
    def __init__(self, p):
        # p is the true success rate 
        self.p = p 
        # a and b are parameters for beta prior distributions 
        self.a = 1 
        self.b = 1 
        
    # draw a sample from the Bernoulli distribution with parameter p 
    def pull(self):
        # sample a random variable in [0.0, 1.0) and compare it with p 
        return np.random.random() < self.p 
    
    # draw a sample from the beta prior distribution 
    def sample(self):
        return np.random.beta(self.a, self.b)
    
    # takes a sample x and update a and b, online learning to update each bandit as soon as we sampled the data
    def update(self, x):
        self.a += x 
        self.b += 1 - x 
```

```python
# plot the posterior distributions 
def plot(bandits, trial):
    x = np.linspace(0, 1, 200)
    for bandit in bandits:
        y = beta.pdf(x, bandit.a, bandit.b)
        plt.plot(x, y, label = 'real p: %.4f' % bandit.p, alpha = 0.5)
        plt.title('Bandit distributions after %s trials' %trial)
        plt.legend()
    plt.show()
```

```python
# experiment methods 
def experiment():
    # a list fo Bandit objects with true success rate given 
    bandits = [Bandit(p) for p in BANDIT_PROBABILITIES]
    
    # iterations where we'd like to pause to make a plot 
    sample_sizes = [5, 10, 20, 50, 100, 200, 500, 1000, 1500, 2000]
    
    # loop over NUM_TRIALS
    for i in range(NUM_TRIALS):
        # initialize variables to store the current best bandit with its corresponding probability from beta 
        best_bandit = None 
        max_score = -1 
        
        all_samples = []
        
        for bandit in bandits:
            # sample with the current beta prior distribution 
            sample = bandit.sample()
            
            all_samples.append("%.4f" %sample)
            if sample > max_score:
                max_score = sample 
                best_bandit = bandit 
        if i in sample_sizes:
            print('Current samples: %s' %all_samples)
            plot(bandits, i)
            
        # pull arm for the bandit with the highest score (ranking)
        x = best_bandit.pull()
        
        # update bandit's distribution with new data 
        best_bandit.update(x)
        
if __name__ == '__main__':
    experiment()
```
